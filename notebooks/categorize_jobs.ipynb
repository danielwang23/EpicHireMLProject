```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/danielwang/EpicHire/Data-Science-Team/data/job_descriptions.csv')

# Limiting to the first 200 job postings for faster analysis
df = df.head(200)


# Define categories and associated keywords
categories = {
    'Technology': ['developer', 'engineer', 'data', 'software', 'it', 'analyst', 'system', 'technician', 'programmer'],
    'Social Science': ['psychologist', 'sociologist', 'social worker', 'counselor', 'researcher'],
    'Life Science': ['biologist', 'chemist', 'medical', 'pharmacist', 'nurse', 'doctor', 'health', 'physician'],
    'Finance': ['accountant', 'finance', 'financial', 'investment', 'analyst', 'banking', 'auditor'],
    'Marketing': ['marketing', 'advertising', 'seo', 'brand', 'public relations', 'communications'],
}

categorized_data = {category: pd.DataFrame(columns=df.columns) for category in categories}


# Function to categorize a job posting
def categorize_posting(row):
    job_title = str(row['Job Title']).lower()
    role = str(row['Role']).lower()
    job_description = str(row['Job Description']).lower()
    
    for category, keywords in categories.items():
        if any(keyword in job_title or keyword in role or keyword in job_description for keyword in keywords):
            categorized_data[category] = categorized_data[category].append(row)
            return category  # Return the category if a match is found
    return 'Uncategorized'  # If no category is matched
```

```python
print("Columns in the original DataFrame:", df.columns.tolist())
```

```python
# Apply the categorization function
df['Category'] = df.apply(categorize_posting, axis=1)

columns_to_display = [
    'Job Id', 'Company', 'Job Title', 'Job Description', 'Qualifications', 'Experience', 
    'Role', 'Responsibilities', 'skills', 'Work Type'
]

# Qualifications: The educational qualifications needed for the job.
# Experience: The required or preferred years of experience for the job.
# Role: The role or category of the job (e.g., software developer, marketing manager).
# Responsibilities: Specific responsibilities and duties associated with the job.
# Skills: The skills or qualifications required for the job.

for category, data in categorized_data.items():
    # Select only columns that are both in `columns_to_display` and the DataFrame `data`
    available_columns = [col for col in columns_to_display if col in data.columns]
    categorized_data[category] = data[available_columns]


    print(f"\nColumns in '{category}' category after filtering:", categorized_data[category].columns.tolist())
```

```python
# Display sample results for each category
for category, data in categorized_data.items():
    print(f"\nSample records for {category} category with selected columns:")
    display(data.head())
```

```python
#Saving each category to a separate CSV file
for category, data in categorized_data.items():
    if not data.empty:
        filename = f'/Users/danielwang/EpicHire/Data-Science-Team/data/{category}_job_postings.csv'
        data.to_csv(filename, index=False)
        print(f"{category} category saved to {filename}")
```
